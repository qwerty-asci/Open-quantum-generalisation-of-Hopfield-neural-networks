```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sco



plt.figure(1,figsize=(8,8))
plt.rcParams["axes.facecolor"] = "#e2ff8f"
#plt.title(r"Diagrama de fase $\Omega-T$ para la solucin paramagntica",fontsize=25)
plt.title(r"Solucin paramagntica",fontsize=25)
plt.xlabel("T",fontsize=25)
plt.ylabel(r"$\Omega$",fontsize=25)

plt.ylim(0.0,1.0)
plt.xlim(0.0,1.2)

x=np.linspace(0.0001,1.0,1000)

def f(x):
    a=np.zeros(x.size)
    for i in range(a.size):
        if(x[i]<=1.0):
            a[i]=np.sqrt((1/x[i]-1)/8)
        else:
            a[i]=0.0
    return a


def g(x):
    a=np.zeros(x.size)
    for i in range(a.size):
        if(x[i]<=1.0):
            a[i]=np.sqrt((1/x[i]-1)/8)
        else:
            a[i]=0.0
    return a

#Primera fase
plt.plot(x,np.sqrt((1/x-1)/8),color="#000000")
plt.fill_between(x, np.sqrt((1/x-1)/8), where=x <= 1 , facecolor="#cdffec",alpha=0.5)

#beta=3/2
plt.plot([2/3,2/3],[0.25,9],linestyle="--",color="#000000")

#Segunda fase:
#Situacin T>=2
x=np.linspace(0.0001,1.19,1000)
#curva inferior
plt.plot(x,np.abs(1/x-0.5)/4.0,color="#000000")
x=np.linspace(0.0001,1.0,1000)
#curva superior
plt.plot(x,np.sqrt((1/x-1)/8),color="#000000")


plt.annotate('III.I',xy=(0.3,0.54),xytext=(0.4,0.3),arrowprops=dict(shrink=0.1,width=0.1,headwidth=4,headlength=3,color="#2b68ff"),fontsize=20)
plt.annotate('III.II',xy=(0.85,0.12),xytext=(1,0.4),arrowprops=dict(shrink=0.1,width=0.1,headwidth=4,headlength=3,color="#2b68ff"),fontsize=20)

plt.fill_between(x, np.abs(1/x-0.5)/4.0,np.sqrt((1/x-1)/8), where=x <= 2/3, facecolor="pink",alpha=0.5)
x=np.linspace(0.0001,2.0,1000)
plt.fill_between(x, np.abs(1/x-0.5)/4.0,f(x), where=x >= 2/3, facecolor="#b3b8fc")
#Caso betac=3/2
#plt.axhline(1/4.0,linestyle="--",color="#000000")

#Caso betac=9/8
#plt.axhline(np.sqrt((9/8-1)/8),linestyle="--")

#tercera fase
x=np.linspace(0.0001,5,1000)
#curva superior
plt.plot(x,np.abs(-1/x+0.5)/4.0,color="#000000")

#ultima fase
plt.fill_between(x,5.0, np.abs(1/x-0.5)/4.0, where=x >= 2/3-0.001, facecolor="#ddcdff",alpha=1)

plt.xticks(fontsize=17)
#plt.legend(loc="upper right",fontsize=17)
plt.yticks(fontsize=17)

#Textos
plt.text(0.4,0.2,"I",fontsize=20)
plt.text(0.18,0.8,"II.I",fontsize=20)
plt.text(1.05,0.03,"II.II",fontsize=20)
plt.text(0.9,0.8,"IV.I",fontsize=20)
plt.text(0.4,0.8,"IV.II",fontsize=20)

plt.savefig("Diagrama_de_fase_solucion_paramagnetica.png",format="png")
```
