```python
import numpy as np
import matplotlib.pyplot as plt

mz=np.genfromtxt("m_totalz_T.dat")

qbits=7

Omega=1.2

beta=1/np.linspace(0.005,1.1,20)

t=np.arange(0,4*mz.shape[0],4)/10000


for i in range(mz.shape[1]):
    plt.figure(i,figsize=(8,6))

    #plt.title(r"m$_z$ para "+str(qbits)+" qbits con $\Omega=$"+str(Omega),fontsize=25)
    plt.title(r"m$_z$ con T="+str(1/beta[i])[0:6]+" y $\Omega=$"+str(Omega),fontsize=25)
    plt.ylim(-1.3,1.3)
    plt.plot(t,mz[:,i],linewidth=1.5,color="#6cd7ff")

    plt.xlim(t[0],t[-1])
    plt.xlabel("t",fontsize=25)
    plt.ylabel(r"m$_{z}$",fontsize=25)
    plt.xticks(fontsize=17)
    #plt.legend(loc="upper right",fontsize=17)
    plt.yticks(fontsize=17)

    plt.savefig("mz_en_funcion_de_t_T_"+str(1/beta[i])[0:6]+"_"+str(qbits)+"_omega_"+str(Omega)+".png",format="png")
```

```python
for i in range(mz.shape[1]):
    plt.figure(i,figsize=(8,6))

    plt.title(r"Histograma de m$_z$ con T="+str(1/beta[i])[0:6]+" y $\Omega=$"+str(Omega),fontsize=22)
    #plt.ylim(-1.1,1.1)
    plt.hist(mz[:,i],bins=100,density=True,histtype='step',color="#6cd7ff")

    plt.xlabel(r"m$_{z}$",fontsize=25)
    plt.xlim(-1,1)
    plt.xticks(fontsize=17)
    plt.yticks([])

    plt.savefig("histograma_T_"+str(1/beta[i])[0:6]+"_"+str(qbits)+"_omega_"+str(Omega)+".png",format="png")
```
